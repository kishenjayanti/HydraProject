## Building a Recommendation Engine Based on User Rating

### 1. Creating Functions to calculate simmilarity

```python
# Fucntion for Calculating Eucledian Distance
def sim_distance(rate, user1, user2):
    
    # Get list of movies watched by both users
    s = {}
    for i in rate[user1]:
        if i in rate[user2]:
            s[i] = 1
            
    # Return 0 if no ratings in common
    if len(s)==0:
        return 0
    
    # Calculate the euclidean distance
    ssquared = sum([pow(rate[user1][i] - rate[user2][i],2) for i in rate [user1] if i in rate[user2]])
    return 1/(1+ssquared)
```

```python
# Function for Calculating Pearson Correlation
def sim_pearson(rate,u1,u2):
    # Get list of movies rated by both users
    s = {}
    for i in rate[u1]:
        if i in rate[u2]:
            s[i] = 1
    # Calculate the number of such movies
    n = len(s)
    
    # Return 0 if no ratings in common
    if n==0: return 0
    
    # Sum of all the ratings
    s1 = sum([rate[u1][t] for t in s])
    s2 = sum([rate[u2][t] for t in s])
    
    # Sum and product of squares of the ratings
    s1sq = sum([pow(rate[u1][t],2) for t in s])
    s2sq = sum([pow(rate[u2][t],2) for t in s])
    psum = sum([rate[u1][t]*rate[u2][t] for t in s])
    
    # Get the pearson correlation score
    num = psum-(s1*s2/n)
    den = sqrt((s1sq - pow(s1,2)/n)*(s2sq-pow(s2,2)/n))
    if den == 0: return 0
    r = num/den
    return r
```

### 2. Extracting data into dictionaries

```python
def loadmovielens(path = "C:\Users\SabariMukunthan.VST-PC\Desktop\ml-20m"):
    movies = {}
    for line in open(path+'\\movies.txt'):
        (id,title,genre) = line.split('\t')
        movies[id] = title
    prefs={}
    for line in open(path+'\\ratings.txt'):
        (userid, movieid, ratings, timestamp) = line.split('\t')
        prefs.setdefault(userid,{})
        prefs[userid][movies[movieid]] = float(ratings)
    return prefs
dataset = loadmovielens()
```

### 3. Building the reccomendation Function

```python
def getrecommendations(prefs,person,similarity = sim_pearson):
    totals = {}
    simsums = {}
    for other in prefs:
        #Ignoring the user
        if other == person:
            continue
        sim = similarity(prefs,person,other)
        # Ignoring score <0
        if sim <=0: continue
        for item in prefs[other]:
        # Only score movies not seen by the user
            if item not in prefs[person] or prefs[person][item] == 0:
                totals.setdefault(item,0)
                totals[item]+= prefs[other][item]*sim
        # Sum of Simmilarities
                simsums.setdefault(item,0)
                simsums[item]+=sim
        # Creating Final List
    rankings = [(total/simsums[item],item)for item,total in totals.items()]
    rankings.sort()
    rankings.reverse()
    return rankings
getrecommendations(dataset,'6589')
```
